```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation 
1. whenever the rating is between 4.5 to 4.9 it is consider as the Excellent.
2. whenever the rating is between 4.0 to 4.4 it is consider as the Very Good.
3. whenever the rating is between 3.5 to 3.9 it is consider as the Good.
4. whenever the rating is between 2.5.5 to 3.4 it is consider as the Average.
5. whenever the rating is between 1.8 to 2.4 it is consider as the Poor.

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,)
```

##color mapping is different here:
so we have to map the color because it is not matching with the rating data 

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['Blue','Red','Orange','Yellow','Green', 'Green'])
```

## Observations :
      1. Not rated count is very high.
      2. Maximum number of rating are lies between the 2.5 - 3.4 .
   


```python
#COUNT PLOT
sns.countplot(x='Rating color',data = ratings,palette=['Blue','Red','Orange','Yellow','Green', 'Green'])
```
